```python
import matplotlib.pyplot as plt
import numpy as np
import math
from matplotlib import animation, rc
rc('animation', html='jshtml') # or 'html5'
```

```python
def y1(x,t):
    return (0.1)*np.exp(-(x-t+4)**2/(2*0.5)**2)+(1/30)*np.exp(-(-x-t+4)**2/(2*0.5)**2)
```

```python
def y2(x,t):
    return (4/30)*np.exp(-(x-2*t+8)**2/((2)**2))
```

```python
def y_tot(x,t):
    return np.where(x<=0, y1(x,t), y2(x,t))
```

```python
tp = np.arange(0,15, 0.15) # [s]
x = np.arange(-10.,20.,0.01) # [m]

fig = plt.figure(figsize=(12, 3))
plt.xlim(-10.,20.)
plt.ylim(0.0,0.2)
plt.xlabel("x [m]")
plt.ylabel("Vertical Displacement (m)")
plt.title(r"Reflection and Transmission of $y(x,t)$ ($t \in [0,10]$)")
plt.axvline(x=0, c="gray")
plt.tight_layout(pad=1.00)
data, = plt.plot([],[])

def init():
  data.set_data(x, y_tot(x,0))
  return data,

def animate(i):
  t = tp[i]
  ytot = y_tot(x,t)
  data.set_data(x, ytot)
  return data,

anim = animation.FuncAnimation(fig, animate, init_func=init, frames=len(tp), interval=50, blit=True)

plt.close(fig)  # This hides the original plot.
anim
```
