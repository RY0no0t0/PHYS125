```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fftfreq
```

```python
piano = np.loadtxt("piano.csv")
```

```python
len(piano)
```

```python
g=np.fft.fft(piano)         #This line applies the fft function to the date points stored in the wave3 array.
N=len(piano)             #This is the total number of samples.
SR=44100           #This number is the sample rate, which is the number of samples recorded per second.
xf =fftfreq(N, 1/SR) #This line applies the fftfreq function to convert the angular frequency values to frequency values.


plt.plot(np.abs(xf), np.abs(g))  #This line plots the Fourier coefficients in terms of frequency values.
                       # The abs function only gives the positive Fourier coefficients.

plt.
plt.show()
```
