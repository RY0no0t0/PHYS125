```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import math
```

```python
data=np.loadtxt("data.csv", delimiter=',', skiprows=1, usecols=(0, 1), dtype=float)
```

```python
# check the data
plt.plot(data[:,0], data[:,1])
```

```python
# define function -- solution to simple harmonic oscillation
def theta (t, A, omega, phi):
    return A*np.sin(omega*t+phi)
```

```python
# convert from degrees to radian
y = data[:,1]/180*math.pi
```

```python
# run curve fit
popt, pcov = curve_fit(theta, data[:,0], y, p0=[0.4, 8.5, 0.1])
```

```python
# show parameters
popt
```

```python
plt.figure(figsize=(12,6))
plt.plot(data[:,0], theta(data[:,0],*popt), color="salmon", label="curve fit")
plt.scatter(data[:,0], y, label="data")

plt.legend()
plt.ylabel(r"$\theta$ (rad)")
plt.xlabel("time (s)")
plt.show()
```
