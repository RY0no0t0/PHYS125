## Question 6

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Define parameters of this question
A = 0.10 #[m]
omega = (0.400/0.0300)**(1/2) #[rad/s], fixed m to [kg] and k is [N/m]
phi = 0 #[rad]
```

```python
# Define an array of time in one period
T = 2.*math.pi/(omega) #[s]

t =  np.linspace(0, T, 30)
```

```python
# defining the eq. of motion
def x(t, A, omega, phi):
    return A*np.cos(omega*t+phi)
```

```python
# plot over one period
plt.plot(t, x(t, A, omega, phi))
plt.ylabel("x(t) [m]")
plt.xlabel("time [s]")
plt.show()
```

## Question 7

```python
# Define parameters of this question
x0 = 0.1 #[m]
omega0 = math.pi #[rad/s]
phi0 = 0 #[rad]
gamma = math.pi/15 #[s^-1]
```

```python
# Define an array of times to plot
t =  np.linspace(0, 15, 200)
```

```python
# defining the eq. of motion
def x(t, x0, omega0, phi0, gamma):
    return x0*np.exp(-gamma*t/2)*np.cos((omega0**2-(gamma**2)/4)**(1/2)*t+phi0)
```

```python
# plot the function
plt.plot(t, x(t, x0, omega0, phi0, gamma))
plt.ylabel("x(t) [m]")
plt.xlabel("time [s]")
plt.show()
```
