```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def irradiance(theta, lam, a, b, I0=1):
    k = 2*math.pi/lam #define k from lam (wavelength)

    alpha = k*a/2*np.sin(theta)
    beta = k*b/2*np.sin(theta)

    return 4*I0*((np.cos(alpha))**2)*((np.sin(beta)/beta)**2)
```

```python
thetas = np.linspace(-math.pi/100, math.pi/100, 1000)
plt.plot(thetas*1500, irradiance(thetas, 0.650, 300, 80))
plt.xlabel("Lateral Displacement from the center of two slits [mm]")
plt.ylabel(r"Irradiance / $I_0$")

plt.show()
```
