## *fft* function

To find the Fourier transform coefficients the computer needs to solve the Fourier integrals by computing it as a discrete sum, which is called the Discrete Fourier transform. The function that does these calculations in Python is called *dft*. The *dft* function is feasible for limited data points. When dealing with large data points dft can take a long time executing the code. A faster function to use is called the Fast Fourier transform (*fft*), which we will use for this activity. The *fft* function can be imported from the *numpy* package.  


To find the Fourier coefficients, we use the following syntax, which applies the *fft* function to an array of input data points, computes their Fourier coefficients and stores them in a new array, with its name written on the left side of the code line as below:



```python
g=fft(y)   #This line applies the fft function to the data points in
           #the array y, computing their Fourier coefficients and stores them in a new array called g.
```

## ifft function

The function that calculates the inverse of Fourier transform is called *ifft*, where letter i represents the inverse. When we apply this function to an array of Fourier coefficients it returns back an array of original datapoints.

By using the syntax below we can find the inverse of Fourier coefficients and store them in a new array called b.

```python
b=ifft(g) #This line calculates the inverse of Fourier coefficients and stores the values in array b.
```

# Example

In the program below, we will apply the *fft* function to the data points stored in the wave3 array to get the Fourier transform coefficients in the angular frequency domain. To change from the angular frequency domain ($k$) to frequency ($f$) domain we will convert k to f by using the *fftfreq* function from *scipy.fft* package.


```python
from google.colab import drive
drive.mount('/content/drive')

from numpy import *
from pylab import *
from matplotlib import *
from scipy.fft import fftfreq

g=fft(wave3)         #This line applies the fft function to the date points stored in the wave3 array.
N=124026             #This is the total number of samples.
SR=22050             #This number is the sample rate, which is the number of samples recorded per second.
xf =fftfreq(N, 1/SR) #This line applies the fftfreq function to convert the angular frequency values to frequency values.


plot(abs(xf), abs(g))  #This line plots the Fourier coefficients in terms of frequency values.
                       # The abs function only gives the positive Fourier coefficients.
show()
```

```python
xlim(0, 2000)
plot(abs(xf), abs(g))
show()
```

From the plot above we can see the main frequency of the whistle is about 1750 Hz.
