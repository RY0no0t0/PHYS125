## *for* loop statement
In programming a *for* loop can be used for recurring calculations.

The syntax of a *for* loop has two parts: The number of iterations that indicate how many times the program needs to go through a loop to perform a task or a certain number of tasks, and the indentation part, which includes a task or a list of tasks to be performed in each round of the loop.

After going through all the required number of iterations the program exits the *for* loop and go to the next line of code. You can find more information on a *for* loop in the Introduction to Python 3 file.



## *if* statement

An *if* statement combined with an *else* statement is used to apply a condition to the program over which certain tasks are executed depending on whether the condition is met, that is, whether the condition is True or False.

The syntax of an *if* statement has two parts, the condition part which ends with a colon (:), followed by the indentation part, which includes the list of task(s).


For example, in the code line below the condition compares the number $x$ with number $1$ and evaluates whether it is larger than $1$. If the value of $x$ is larger than $1$, the condition is met and the program goes into the indentation  part and the *print* statement gets executed. If $x$ value is not larger than $1$ the program does not go through the indentation part and goes to the next code line.

```python
x=10
if x>1:
  print('The value of x is larger than 1.')
```

If the indentation part becomes aligned with the condition part of the *if* statement the program gets an error as in the example below.


```python
x=10
if x>1:
print('The value of x is larger than 1.')
```

## Example  

In the example below, we use a *for* loop and an *if* statement to write a code to plot the piecewise function
$y_{\rm tot} = \left\{\begin{array} xx &, x\le 0\\ \sin(x) &, x>0\end{array}\right.$

By adding the *for* loop, Python goes through each $x$ value in the list created by the *linspace* statement. Inside the *for* loop, the *if* statement in each iteration compares the $x$ value with the interface value of $x$, which is defined at the beginning of the program.

The *if* statement stores the corresponding $x$ and $y$ values in their corresponding lists. For that we have defined empty lists at the beginning of the program with the same length as the length of the $x$ list.

After iterating through all the $x$ values the program exits the *for* loop and goes through the next line, which is the *plot* statement. The *plot* statement plots the stored $y$ values as a function of stored $x$ values.  



```python
from numpy import *
from pylab import plot, show

N=30  # N is the length of the x list.
x=zeros(N)
y=zeros(N)
x=linspace(-5, 5, 30) # The x points are defined over a range of (-5, 5).

x_int=0               # Defining an interface point.

def y_tot(x):
    for i in range(N):         # Loop through all x values.
        if x[i]<=x_int:        # In each iteration, the i^th x value is compared to the x_int value.
            y[i]=x[i]          # When the if statement condition is true this indented code gets executed.
        else:
            y[i]=sin(x[i])     # When the else statement condition is true this indented code gets executed.
    return y

plot(x, y_tot(x))
show()
```
