##Introduction to Plotting in Python


###Import functions from pylab
Let's say the goal is to plot sin(θ) as a function of some independent variables such as different angles (θ).
For that we need to define a range of angles for example between 0 and 2π.


###linspace syntax
In order to define a range of values in Python we can use the *linspace* statement from the *numpy* package. The *linespace* will store all of the values of a range in an array. The format of *linespace* is linspace(starting point, ending point, number of divisions). For example, to define a range of angles between 0 and 2π we set the starting point as zero, the ending point as 2π and number of divisions as 20 with the following syntax.

theta=linspace(0, 2*pi, 20)
theta is a variable which has stored an array of all the 20 numbers in the given range defined by the *linspace* statement. We can create more data points in a given range by increasing the number of divisions.  

###Applying a function to the datapoints in a linespace
We also need to import the *sin* function from the *numpy* package. Importing the sine function from the *numpy* package makes it an array which can be applied to the range of independent variables already defined by the *linspace* statement. To save these values as a new array you can assign those values to a new variable y such as:

y=sin(theta)

If you would like to see the values of the sine function in the given range you can use the *print* statement such as print(sin(theta)) or print(y).

###*plot* statement syntax

In our example, we are going to use the *plot* statement to plot the dependent values of sin(θ) as a function of the independent variables of (theta).
For that we type plot(theta, sin(theta)). We also need to import another function, which is called *show*. So we need to add this line to the beginning of the code cell too. To import more than one function from the same package we can separate the functions by commas and import them. To perform the plot we can use show() syntax after the plot line.

Below you can see a code cell for plotting a sine function in the range of 0, 2π.

Can you change one number in the code to make the plot less jagged?

```python
from numpy import sin, linspace, pi
from pylab import plot, show

theta=linspace(0, 2*pi, 20)
plot(theta, sin(theta))
show()
```

###Plot format


In Python you can format your plot in different ways such as titling it, titling the axes, using legends, or changing the line style or color coding it.



####Adding a title to a plot

For adding a title to a plot, we need to import *title* from the *pylab* package, using the following syntax:
title(‘sin(theta) v.s. theta’).

You can type the name of your plot inside a single quotation mark. In this example, I have titled my plot as sin(theta) v.s. theta.

####Adding axes titles to a plot

For titling the x axis you need to import *xlabel* from *pylab* package and for titling the y axis you need to import *ylabel* from the *pylab* package.

For adding title to both axes the syntax is:

xlabel(‘x’)
ylabel(‘y’)

In this example, I have titled my x axis as x and the y axis as y.

####Adding legends to a plot

If you would like to add a legend to your plot or have two graphs in one plot and would like to distinguish them by adding legends you can add label to your *plot* statement. For that you will need to import the *legend* function from the *pylab* package.

plot(x1, y1, label=‘y1 v.s. x1’)

legend()

The first code line will plot the y1 data points as a function of x1 data points and label the legend of the plot as y1 v.s. x1.
The second code line will display the legend on the plot.


#### Color coding a plot
You can also change color of a plot by indicating the color in the *plot* statement using the following syntax:

plot(x1, y1, label=‘ y1 v.s. x1’, color='green')

This code line will change the plot's color to green.
You can see these code lines in the program below:


```python
from numpy import sin, linspace
from math import pi
from pylab import plot, show, title, xlabel, ylabel, legend

theta=linspace(0, 2*pi, 20)
plot(theta, sin(theta), label='sin(theta) v.s. theta', color='green')

title('sin(theta) v.s. theta(radian)')
xlabel('theta (radian)')
ylabel('sin(theta)')

legend()
show()
```

#### Greek letters
It's possible to include Greek symbols like $\theta$ directly in the plot. To do so, you need to preceed the quoted text with the letter r (meaning "raw"), preceed the Greek letter's name with a backslash \\ and surround it with dollar signs. For example, rather than title(‘sin(theta) v.s. theta’), you would enter title(r‘sin(\$\\theta\$) v.s. \$\\theta\$’).

```python
from numpy import sin, linspace
from math import pi
from pylab import plot, show, title, xlabel, ylabel, legend

theta=linspace(0, 2*pi, 20)
plot(theta, sin(theta), label=r'sin($\theta$) v.s. $\theta$', color='green')

title(r'sin($\theta$) v.s. $\theta$ (radian)')
xlabel(r'$\theta$ (radian)')
ylabel(r'sin($\theta$)')

legend()
show()
```

####ylim and xlim statements

You can use *xlim*/*ylim* statements to increase or decrease the length of the x axis/y axis to a specific range in your plot.
For that you will need to import these functions from the *pylab* package. The syntax is:

xlim(lower limit, upper limit)

xlim(lower limit, upper limit)


The first code line will start the x axis with whatever number that is typed instead of the lower limit and the axis will end at whatever number that is typed instead of the upper limit in the *xlim* statement.




###Plotting data from a csv file

You can import your own data such as experimental data to Python by different methods. One way is by using the *loadtxt* statement. The *loadtxt* function needs to be imported from the *numpy* package.

You will also need to organize your data into different columns in a file with a csv format. For example, if you want to plot a dependent variable as a function of an independent variable then place the independent variable values in one column and put the dependent variable values in another column.

Once you prepare your csv file, place it somewhere in your Google drive. In your program, type file_path and set that equal to the address of the location where you have placed the csv file such as in the line below:

file_path= '/content/drive/MyDrive/data/periodtable.csv'

The example above suggests that my csv file, which is called periodtable, is in a folder called data inside my Google drive.

In the next line use the *loadtxt* statement to import your data into Python. Choose a name for your data and set that equal to the *loadtxt* statement as below:

data1=loadtxt(file_path, delimiter=',', skiprows=0, usecols=(0, 1), dtype=float)

The *loadtxt* statement in the line above uses the address of the *file_path* statement to import the data into Python. I have named my data as data1. It also uses the first and second columns of the csv file for data analysis. loadtxt will store the data points in a two dimensional array.

In the loadtxt format, by default the first column of the csv file will be indexed as 0 and the second column of the csv file will be indexed as 1. If you have a title in your csv file you can tell loadtxt to skip that by setting skiprows equal to the number of the title's row. If you do not have a title you can type skiprows=0. I also have defined the data type as float numbers.

Next, we need to determine which column will be considered as the independent variable and which column will be considered as the dependent variable. I have chosen the first column in my csv file, which is indexed as 0 in the *loadtxt* statement, to be my x values and the second column in my csv file, which is indexed by 1 in the *loadtxt* statement, to be my y values. I want all of the rows of each column to be included. The following syntax tells Python to use all of the rows.

x=data1[:,0]

y=data1[:,1]

As an example, if I only need to use the first 500 rows of both columns I will use the following syntax:

x=data1[0:500,0]

y=data1[0:500,1]

Now, I am ready to plot the y data as a function of x data using the *plot* statement with its syntax: plot(x, y)
