```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
data = np.array([[87.9, 82.8, 77.7, 72.5, 67.4, 62.2, 56.9, 51.7, 46.4, 40.7, 34.8,28.1],
                 [12.1, 12.2, 12.3, 12.5, 12.6, 12.8, 13.1, 13.3, 13.6, 14.3, 15.2, 16.9]])
```

```python
x = data[0]+data[1]
y = data[0]*data[1]
```

```python
def model(x, m, b):
    return m*x+b
```

```python
popt, pcov = curve_fit(model, x, y, p0=[10.7, 8.18])
```

```python
plt.scatter(x, y)
plt.plot(x, model(x, *popt), label=rf"$f=({popt[0]:.2f} \pm {pcov[0,0]**(1/2):.3f})$ cm")
plt.legend()
plt.title(r"$(s \times s')=f \; (s+s')$ model fit to data")
plt.xlabel(r"$s + s'$ [$cm$]")
plt.ylabel(r"$s \times s'$ [$cm^2$]")
plt.show()
```
