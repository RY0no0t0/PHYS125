```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# # This is the location of the theory points
# file_path1= '/content/drive/MyDrive/data/periodtable.csv'

# # This is the location of your data
# file_path2= '/content/drive/MyDrive/data/Lab2Data.csv'

# If your files have headers, "skiprows" says how many rows to skip
# usecols says use the first and second columns (numbered starting at zero)
model = np.loadtxt("periodtable.csv", delimiter=',', skiprows=1, usecols=(0, 1), dtype=float)

data = np.loadtxt("Lab2Data.csv", delimiter=',', skiprows=1, usecols=(0, 1), dtype=float)
```

```python
fig, ax = plt.subplots()
ax.axhline(y=1, linestyle="--", color="gray")
ax.plot(model[:,0], model[:,1], label="Model")
ax.scatter(data[:,0], data[:,1], color="mediumvioletred", label="data")

plt.legend()
plt.xlabel(r'$\theta_0$ (radians)')
plt.ylabel(r'$T/T_0$')
plt.ylim(top=3)
plt.title("The error in small-angle approximated period vs amplitude")
plt.show()
```
