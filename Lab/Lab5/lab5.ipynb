```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
data = np.array([[16.3, 24.3, 32.62, 40.78],[0.76*2,0.52*2, 0.36*2, 0.625]])
```

```python
def linear(x, m, b):
    return m*x+b
```

```python
popt, pcov = curve_fit(linear, data[0]**(-1), data[1], p0=[18, 0.5])
```

```python
(pcov[1][1])**(1/2)
```

```python
plt.plot((data[0])**(-1), data[1], label="data")
plt.plot((data[0])**(-1), linear(data[0]**(-1), *popt), label=f"m={popt[0]:.3}, b={popt[1]:.2}")
plt.xlabel("Period (s)")
plt.ylabel("Wavelength (m)")
plt.legend()
plt.title("Transverse Motion")
plt.show()
```

```python
data2 = np.array([[27.5,  34.0, 40.5, 47.2],[0.243*2,0.40,0.34, 0.28]])
```
