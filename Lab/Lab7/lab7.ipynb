```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
#y and y'
y = 1.5 #cm
y_prime = np.array([0.75, 1.6, 2.7])
```

```python
#S and S'
data = np.array([[24.1-8.3, 21.3-11., 22.95-14.6],
                 [8.3, 11., 14.6]])
```

```python
x = data[0]+data[1]
y = data[0]*data[1]
```

```python
def model(x, m, b):
    return m*x+b
```

```python
popt, pcov = curve_fit(model, x, y, p0=[10.7, 8.18])
```

```python
plt.scatter(x, y)
plt.plot(x, model(x, *popt), label=rf"$f=({popt[0]:.2f} \pm {pcov[0,0]**(1/2):.3f})$ cm")
plt.legend()
plt.title(r"$(s \times s')=f \; (s+s')$ model fit to data")
plt.xlabel(r"$s + s'$ [$cm$]")
plt.ylabel(r"$s \times s'$ [$cm^2$]")
plt.show()
```

```python
f = 6.29
d = 20.
s = 1.8
p = 8
s1 = 1/(1/f-1/s)
```
